```python
def outlier(data):
    anamolies = []
    random_state_std = np.std(data)
    random_state_mean = np.mean(data)
    anamoly_factor  = random_state_std * 3
    
    upper_limit = random_state_mean + anamoly_factor
    lower_limit = random_state_mean - anamoly_factor
    
    print(upper_limit)
    print(lower_limit)
    
    for i in data:
        if i > upper_limit or i < lower_limit:
            anamolies.append(i)
#            print(len(anamolies))
    return anamolies

outlier(dodge['length_of_service'])

```

```python
outlier(dodge['age'])
```
